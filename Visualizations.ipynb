```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
moviedf= pd.read_csv("Resources/full_table_rotten_analysis.csv")
moviedf.head(50)
```

```python
moviedf.columns.tolist()
```

```python
movie_by_month_df=moviedf.groupby('release_month')
movie_by_month_df.head()
```

```python
#create moviemonth dataframe
jan_movie = moviedf[(moviedf["release_month"] == 1)]
feb_movie = moviedf[(moviedf["release_month"] == 2)]
mar_movie = moviedf[(moviedf["release_month"] == 3)]
apr_movie = moviedf[(moviedf["release_month"] == 4)]
may_movie = moviedf[(moviedf["release_month"] == 5)]
jun_movie = moviedf[(moviedf["release_month"] == 6)]
jul_movie = moviedf[(moviedf["release_month"] == 7)]
aug_movie = moviedf[(moviedf["release_month"] == 8)]
sep_movie = moviedf[(moviedf["release_month"] == 9)]
oct_movie = moviedf[(moviedf["release_month"] == 10)]
nov_movie = moviedf[(moviedf["release_month"] == 11)]
dec_movie = moviedf[(moviedf["release_month"] == 12)]
jan_movie.head()
```

```python
#group datafram by month and find average rotten tomato score
jan_score = jan_movie.mean()["tomatometer_rating"]
feb_score = feb_movie.mean()["tomatometer_rating"]
mar_score = mar_movie.mean()["tomatometer_rating"]
apr_score = apr_movie.mean()["tomatometer_rating"]
may_score = may_movie.mean()["tomatometer_rating"]
jun_score = jun_movie.mean()["tomatometer_rating"]
jul_score = jul_movie.mean()["tomatometer_rating"]
aug_score = aug_movie.mean()["tomatometer_rating"]
sep_score = sep_movie.mean()["tomatometer_rating"]
oct_score = oct_movie.mean()["tomatometer_rating"]
nov_score = nov_movie.mean()["tomatometer_rating"]
dec_score = dec_movie.mean()["tomatometer_rating"]
```

```python
avg_rat_per_month= {'Month':['Jan','Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
                   'Rotten_Tomato_Score': [jan_score, feb_score, mar_score, apr_score, may_score, jun_score, jul_score, aug_score, sep_score, oct_score, nov_score, dec_score]}
avg_rat_per_month_df=pd.DataFrame(avg_rat_per_month)
avg_rat_per_month_df.head(12)
```

```python
#Set bar chart axis
x_axis = avg_rat_per_month_df["Month"]
y_axis = avg_rat_per_month_df["Rotten_Tomato_Score"]
col = []
label= ["Rotten", "Fresh"]
for lab in label:
    if lab == "Rotten":
        color = 'green'
    else:
        color= 'red'
for val in y_axis:
    if val < 60:
        col.append('green')
    else:
        col.append('red')
plt.bar(x_axis, y_axis, color = col, label= 'Fresh')
plt.xlabel("Month")
plt.ylabel("Rotten Tomato Score")
plt.title("Average Tomatometer Status by Month")
plt.ylim(50,70)
plt.legend()
                     
```

```python
#make list where audience and critic scores align
critics_audience_agree=[]
critic_score= moviedf['tomatometer_status']
audience_score= moviedf['audience_status']

    
```

```python
critic_score_val=[]
audience_score_val=[]
for i in critic_score:
    if i == 'Fresh':
        critic_score_val.append(1)
    else:
        critic_score_val.append(0)
for i in audience_score:
    if i == 'Upright':
        audience_score_val.append(1)
    else:
        audience_score_val.append(0)
```

```python
moviedf["critic_score_val"]= critic_score_val
moviedf["audience_score_val"]=audience_score_val
moviedf.head()
```

```python
#define conditions
conditions =[((moviedf['critic_score_val']==1) & (moviedf['audience_score_val']==1)), 
             ((moviedf['critic_score_val']==1) & (moviedf['audience_score_val']==0)),
             ((moviedf['critic_score_val']==0) & (moviedf['audience_score_val']==1)),
             ((moviedf['critic_score_val']==0) & (moviedf['audience_score_val']==0))]

#define choices
choices = ['Yes', 'No','No','Yes']

#create new column in DataFrame that displays results of comparisons
moviedf['critics_audience_agree'] = np.select(conditions, choices)

moviedf.head()
```

```python
moviedf.info()
```

```python
critics_agree=moviedf[moviedf['critics_audience_agree']=='Yes'].critics_audience_agree.count()
critics_disagree=moviedf[moviedf['critics_audience_agree']=='No'].critics_audience_agree.count()
```

```python
#find sums of each
y_axis=[critics_agree, critics_disagree]
x_axis=["Audience Agree", "Audience Disagree"]
explode = (.2,0)
colors = ("green", "gold")
plt.pie(y_axis, explode=explode, labels= x_axis, autopct='%.1f%%', colors=colors)
plt.title("Pertage Critic Review Aligns\n" + "With Audience Review", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```
