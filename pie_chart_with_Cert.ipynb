```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv ('rotten_tomatoes_movies.csv')
df
reviews = df['tomatometer_status'].combine('Fresh',)
reviews= reviews.value_counts()
labels =["Fresh", "Rotten", "Certified Fresh"]

colors= ["#d62728","#458B00","#FFD700"]
plt.pie(reviews, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=296)
plt.title("Rotten Tomatoes: Fresh vs Rotten")
plt.tight_layout()
```
