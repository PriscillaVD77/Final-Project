```python
#import dependencies
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
from sqlalchemy import create_engine
import psycopg2
```

```python
#read in dataframe
moviedf= pd.read_csv("../rotten_tomatoes_movies.csv")
moviedf.head(10)
```

```python
#drop uneeded columm
moviedf= pd.DataFrame(moviedf)
moviedf = moviedf.drop('critics_consensus', 1)
moviedf
```

```python
#drop NAN
moviedf= moviedf.dropna()
moviedf.head()
```

```python
moviedf.shape
```

```python
#change certified fresh to fresh status
moviedf["tomatometer_status"] = moviedf["tomatometer_status"].replace("Certified-Fresh","Fresh")
moviedf.tomatometer_status.value_counts()
```

```python
#make new columns with release year
moviedf['release_year'] = pd.DatetimeIndex(moviedf['original_release_date']).year
moviedf['streaming_release_year'] = pd.DatetimeIndex(moviedf['streaming_release_date']).year
moviedf.head()
```

```python
#make new columns with release month
moviedf['release_month'] = pd.DatetimeIndex(moviedf['original_release_date']).month
moviedf['streaming_month'] = pd.DatetimeIndex(moviedf['streaming_release_date']).month
moviedf.head()
```

```python
#drop date columns
moviedf = moviedf.drop(['original_release_date', 'streaming_release_date'], 1)
moviedf.head()
```

```python
#make a table for pre production features
movie_info_df= moviedf[['rotten_tomatoes_link','movie_title','movie_info','content_rating','genres','directors', 'authors', 'actors', 'runtime', 'production_company', 'release_year','streaming_release_year', 'release_month', 'streaming_month']]
movie_info_df.head()
```

```python
#make a table for post production df
rating_info_df= moviedf[['rotten_tomatoes_link','tomatometer_status','tomatometer_rating','tomatometer_count','audience_status','audience_rating','audience_count', 'tomatometer_top_critics_count', 'tomatometer_fresh_critics_count','tomatometer_rotten_critics_count']]
rating_info_df.head()
```

```python
#find datatypes of tables
movie_info_df.info()
rating_info_df.info()
```

```python
#export csv files
movie_info_df.to_csv('Resources/movie_info.csv', encoding='utf-8', index=False)
```

```python
#export csv files
rating_info_df.to_csv('Resources/rating_info.csv',encoding='utf-8', index=False)
```
